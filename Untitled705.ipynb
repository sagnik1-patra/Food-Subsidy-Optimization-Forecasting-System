```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# =========================================
# 📁 PATHS
# =========================================
csv_path = r"C:\Users\NXTWAVE\Downloads\Food Subsidy Optimization & Forecasting System\Dataset.csv"
save_dir = r"C:\Users\NXTWAVE\Downloads\Food Subsidy Optimization & Forecasting System"
os.makedirs(save_dir, exist_ok=True)

# =========================================
# 📊 LOAD & CLEAN DATA
# =========================================
df = pd.read_csv(csv_path)

for col in df.columns:
    df[col] = (
        df[col].astype(str)
        .str.replace(",", "")
        .str.replace("₹", "")
        .str.strip()
    )
    df[col] = pd.to_numeric(df[col], errors='coerce')

df = df.dropna(axis=1, how='all')
df = df.fillna(df.mean())

if df.shape[1] < 2:
    raise Exception("❌ Need at least 2 numeric columns")

# =========================================
# 🎯 TARGET
# =========================================
target = df.columns[-1]
X = df.drop(columns=[target])
y = df[target]

# =========================================
# 🔀 SPLIT (SAFE)
# =========================================
test_size = 0.5 if len(df) <= 5 else 0.2

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=test_size, random_state=42
)

# =========================================
# ⚙️ SCALING
# =========================================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# =========================================
# 🧬 FITNESS FUNCTION
# =========================================
def fitness(position):
    try:
        position = np.array(position)
        X_w = X_train_scaled * position

        model = RandomForestRegressor(n_estimators=50, random_state=42)
        model.fit(X_w, y_train)

        preds = model.predict(X_test_scaled * position)
        score = r2_score(y_test, preds)

        if np.isnan(score) or np.isinf(score):
            return -1e6

        return score
    except:
        return -1e6

# =========================================
# 🐺 GWOA PARAMETERS
# =========================================
num_wolves = 10
num_features = X_train.shape[1]
iterations = 20

# Initialize wolves' positions
positions = np.random.rand(num_wolves, num_features)
fitness_vals = np.array([fitness(p) for p in positions])

# Alpha, Beta, Delta wolves
alpha_position = positions[np.argmax(fitness_vals)]
alpha_score = fitness_vals[np.argmax(fitness_vals)]

beta_position = np.zeros_like(alpha_position)
beta_score = -np.inf

delta_position = np.zeros_like(alpha_position)
delta_score = -np.inf

# =========================================
# 🐺 GWOA LOOP
# =========================================
for itr in range(iterations):
    a = 2 - itr * (2 / iterations)

    for i in range(num_wolves):
        r1 = np.random.rand()
        r2 = np.random.rand()

        A = 2 * a * r1 - a
        C = 2 * r2

        # Update wolf positions
        D_alpha = np.abs(C * alpha_position - positions[i])
        D_beta = np.abs(C * beta_position - positions[i])
        D_delta = np.abs(C * delta_position - positions[i])

        new_position = (alpha_position - A * D_alpha) + (beta_position - A * D_beta) + (delta_position - A * D_delta)
        new_position = np.clip(new_position, 0, 1)

        new_fitness = fitness(new_position)

        # Update fitness and best positions
        if new_fitness > fitness_vals[i]:
            positions[i] = new_position
            fitness_vals[i] = new_fitness

        if new_fitness > alpha_score:
            delta_position = beta_position
            delta_score = beta_score
            beta_position = alpha_position
            beta_score = alpha_score
            alpha_position = new_position
            alpha_score = new_fitness
        elif new_fitness > beta_score:
            delta_position = beta_position
            delta_score = beta_score
            beta_position = new_position
            beta_score = new_fitness
        elif new_fitness > delta_score:
            delta_position = new_position
            delta_score = new_fitness

    print(f"Iteration {itr+1} Best R2: {alpha_score:.4f}")

print("\n🏆 GWOA Best Score:", alpha_score)

# =========================================
# 🛟 FALLBACK
# =========================================
if alpha_position is None:
    alpha_position = np.ones(num_features)

# =========================================
# 🌳 FINAL MODEL
# =========================================
X_train_gwoa = X_train_scaled * alpha_position
X_test_gwoa = X_test_scaled * alpha_position

model = RandomForestRegressor(n_estimators=200, random_state=42)
model.fit(X_train_gwoa, y_train)

preds = model.predict(X_test_gwoa)

# =========================================
# 📈 METRICS
# =========================================
def safe_metrics(y_true, y_pred):
    try:
        return {
            "R2": float(r2_score(y_true, y_pred)),
            "RMSE": float(np.sqrt(mean_squared_error(y_true, y_pred))),
            "MAE": float(mean_absolute_error(y_true, y_pred))
        }
    except:
        return {"R2": 0, "RMSE": 0, "MAE": 0}

metrics = safe_metrics(y_test, preds)

print("\n📊 GWOA Metrics:", metrics)

# =========================================
# 💾 SAVE FILES
# =========================================
joblib.dump(model, os.path.join(save_dir, "gwoa_model.pkl"))
joblib.dump(scaler, os.path.join(save_dir, "gwoa_scaler.pkl"))

pd.DataFrame([metrics]).to_csv(os.path.join(save_dir, "gwoa_results.csv"), index=False)

pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": preds
}).to_csv(os.path.join(save_dir, "gwoa_predictions.csv"), index=False)

# =========================================
# 📊 GRAPHS
# =========================================

# Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(), annot=True)
plt.title("GWOA Heatmap")
plt.savefig(os.path.join(save_dir, "gwoa_heatmap.png"))
plt.show()

# Accuracy
plt.figure()
plt.bar(["R2"], [metrics["R2"]])
plt.title("GWOA Accuracy")
plt.savefig(os.path.join(save_dir, "gwoa_accuracy_graph.png"))
plt.show()

# Comparison
plt.figure()
plt.bar(["RMSE", "MAE"], [metrics["RMSE"], metrics["MAE"]])
plt.title("GWOA Error Comparison")
plt.savefig(os.path.join(save_dir, "gwoa_comparison_graph.png"))
plt.show()

# Prediction
plt.figure()
plt.plot(y_test.values, label="Actual")
plt.plot(preds, label="Predicted")
plt.legend()
plt.title("GWOA Prediction")
plt.savefig(os.path.join(save_dir, "gwoa_prediction_graph.png"))
plt.show()

# Result graph
plt.figure()
plt.bar(metrics.keys(), metrics.values())
plt.title("GWOA Result Metrics")
plt.savefig(os.path.join(save_dir, "gwoa_result_graph.png"))
plt.show()

# =========================================
# 📄 JSON + YAML
# =========================================
with open(os.path.join(save_dir, "gwoa_results.json"), "w") as f:
    json.dump(metrics, f, indent=4)

with open(os.path.join(save_dir, "gwoa_config.yaml"), "w") as f:
    yaml.dump({
        "target": target,
        "features": list(X.columns),
        "GWOA_best_score": float(alpha_score)
    }, f)

print("\n✅ GWOA PIPELINE COMPLETED SUCCESSFULLY")
```
