```python
# =========================================
# 📦 IMPORTS
# =========================================
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# =========================================
# 📁 PATHS
# =========================================
csv_path = r"C:\Users\NXTWAVE\Downloads\Food Subsidy Optimization & Forecasting System\Dataset.csv"
save_dir = r"C:\Users\NXTWAVE\Downloads\Food Subsidy Optimization & Forecasting System"
os.makedirs(save_dir, exist_ok=True)

# =========================================
# 📊 LOAD & CLEAN DATA
# =========================================
df = pd.read_csv(csv_path)

for col in df.columns:
    df[col] = (
        df[col].astype(str)
        .str.replace(",", "")
        .str.replace("₹", "")
        .str.strip()
    )
    df[col] = pd.to_numeric(df[col], errors='coerce')

df = df.dropna(axis=1, how='all')
df = df.fillna(df.mean())

if df.shape[1] < 2:
    raise Exception("❌ Need at least 2 numeric columns")

# =========================================
# 🎯 TARGET
# =========================================
target = df.columns[-1]
X = df.drop(columns=[target])
y = df[target]

# =========================================
# 🔀 SPLIT (SAFE)
# =========================================
test_size = 0.5 if len(df) <= 5 else 0.2

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=test_size, random_state=42
)

# =========================================
# ⚙️ SCALING
# =========================================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# =========================================
# 🧬 PSO FITNESS FUNCTION
# =========================================
def fitness(position):
    try:
        position = np.array(position)
        X_w = X_train_scaled * position

        model = RandomForestRegressor(n_estimators=50, random_state=42)
        model.fit(X_w, y_train)

        preds = model.predict(X_test_scaled * position)
        score = r2_score(y_test, preds)

        if np.isnan(score) or np.isinf(score):
            return -1e6

        return score
    except:
        return -1e6

# =========================================
# 🧬 PSO PARAMETERS
# =========================================
num_particles = 10
num_features = X_train.shape[1]
iterations = 20

w = 0.5     # inertia
c1 = 1.5    # cognitive
c2 = 1.5    # social

# Initialize particles
positions = np.random.rand(num_particles, num_features)
velocities = np.random.rand(num_particles, num_features)

pbest = positions.copy()
pbest_scores = np.array([fitness(p) for p in positions])

gbest_idx = np.argmax(pbest_scores)
gbest = pbest[gbest_idx]
gbest_score = pbest_scores[gbest_idx]

# =========================================
# 🧬 PSO LOOP
# =========================================
for itr in range(iterations):
    for i in range(num_particles):

        r1 = np.random.rand(num_features)
        r2 = np.random.rand(num_features)

        velocities[i] = (
            w * velocities[i]
            + c1 * r1 * (pbest[i] - positions[i])
            + c2 * r2 * (gbest - positions[i])
        )

        positions[i] = positions[i] + velocities[i]
        positions[i] = np.clip(positions[i], 0, 1)

        score = fitness(positions[i])

        if score > pbest_scores[i]:
            pbest[i] = positions[i]
            pbest_scores[i] = score

    best_idx = np.argmax(pbest_scores)
    if pbest_scores[best_idx] > gbest_score:
        gbest = pbest[best_idx]
        gbest_score = pbest_scores[best_idx]

    print(f"Iteration {itr+1} Best R2: {gbest_score:.4f}")

print("\n🏆 PSO Best Score:", gbest_score)

# =========================================
# 🌳 FINAL MODEL
# =========================================
if gbest is None:
    gbest = np.ones(num_features)

X_train_pso = X_train_scaled * gbest
X_test_pso = X_test_scaled * gbest

model = RandomForestRegressor(n_estimators=200, random_state=42)
model.fit(X_train_pso, y_train)

preds = model.predict(X_test_pso)

# =========================================
# 📈 METRICS
# =========================================
def safe_metrics(y_true, y_pred):
    try:
        return {
            "R2": float(r2_score(y_true, y_pred)),
            "RMSE": float(np.sqrt(mean_squared_error(y_true, y_pred))),
            "MAE": float(mean_absolute_error(y_true, y_pred))
        }
    except:
        return {"R2": 0, "RMSE": 0, "MAE": 0}

metrics = safe_metrics(y_test, preds)

print("\n📊 PSO Metrics:", metrics)

# =========================================
# 💾 SAVE FILES
# =========================================
joblib.dump(model, os.path.join(save_dir, "pso_model.pkl"))
joblib.dump(scaler, os.path.join(save_dir, "pso_scaler.pkl"))

pd.DataFrame([metrics]).to_csv(os.path.join(save_dir, "pso_results.csv"), index=False)

pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": preds
}).to_csv(os.path.join(save_dir, "pso_predictions.csv"), index=False)

# =========================================
# 📊 GRAPHS
# =========================================

# Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(), annot=True)
plt.title("PSO Heatmap")
plt.savefig(os.path.join(save_dir, "pso_heatmap.png"))
plt.show()

# Accuracy
plt.figure()
plt.bar(["R2"], [metrics["R2"]])
plt.title("PSO Accuracy")
plt.savefig(os.path.join(save_dir, "pso_accuracy_graph.png"))
plt.show()

# Comparison
plt.figure()
plt.bar(["RMSE", "MAE"], [metrics["RMSE"], metrics["MAE"]])
plt.title("PSO Error Comparison")
plt.savefig(os.path.join(save_dir, "pso_comparison_graph.png"))
plt.show()

# Prediction
plt.figure()
plt.plot(y_test.values, label="Actual")
plt.plot(preds, label="Predicted")
plt.legend()
plt.title("PSO Prediction")
plt.savefig(os.path.join(save_dir, "pso_prediction_graph.png"))
plt.show()

# Result graph
plt.figure()
plt.bar(metrics.keys(), metrics.values())
plt.title("PSO Result Metrics")
plt.savefig(os.path.join(save_dir, "pso_result_graph.png"))
plt.show()

# =========================================
# 📄 JSON + YAML
# =========================================
with open(os.path.join(save_dir, "pso_results.json"), "w") as f:
    json.dump(metrics, f, indent=4)

with open(os.path.join(save_dir, "pso_config.yaml"), "w") as f:
    yaml.dump({
        "target": target,
        "features": list(X.columns),
        "PSO_best_score": float(gbest_score)
    }, f)

print("\n✅ PSO PIPELINE COMPLETED SUCCESSFULLY")
```
