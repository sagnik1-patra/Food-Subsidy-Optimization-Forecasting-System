```python
# =========================================
# 📦 IMPORTS
# =========================================
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# =========================================
# 📁 PATHS
# =========================================
csv_path = r"C:\Users\NXTWAVE\Downloads\Food Subsidy Optimization & Forecasting System\Dataset.csv"
save_dir = r"C:\Users\NXTWAVE\Downloads\Food Subsidy Optimization & Forecasting System"
os.makedirs(save_dir, exist_ok=True)

# =========================================
# 📊 LOAD & CLEAN DATA
# =========================================
df = pd.read_csv(csv_path)

for col in df.columns:
    df[col] = (
        df[col].astype(str)
        .str.replace(",", "")
        .str.replace("₹", "")
        .str.strip()
    )
    df[col] = pd.to_numeric(df[col], errors='coerce')

df = df.dropna(axis=1, how='all')
df = df.fillna(df.mean())

if df.shape[1] < 2:
    raise Exception("❌ Need at least 2 numeric columns")

# =========================================
# 🎯 TARGET
# =========================================
target = df.columns[-1]
X = df.drop(columns=[target])
y = df[target]

# =========================================
# 🔀 SPLIT
# =========================================
test_size = 0.5 if len(df) <= 5 else 0.2

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=test_size, random_state=42
)

# =========================================
# ⚙️ SCALING
# =========================================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# =========================================
# 🧬 FITNESS FUNCTION
# =========================================
def fitness(position):
    try:
        position = np.array(position)
        X_w = X_train_scaled * position

        model = RandomForestRegressor(n_estimators=50, random_state=42)
        model.fit(X_w, y_train)

        preds = model.predict(X_test_scaled * position)
        score = r2_score(y_test, preds)

        if np.isnan(score) or np.isinf(score):
            return -1e6

        return score
    except:
        return -1e6

# =========================================
# 🦐 KRILL HERD PARAMETERS
# =========================================
num_krill = 10
num_features = X_train.shape[1]
iterations = 20

positions = np.random.rand(num_krill, num_features)
fitness_vals = np.array([fitness(p) for p in positions])

best_idx = np.argmax(fitness_vals)
best_solution = positions[best_idx]
best_score = fitness_vals[best_idx]

# =========================================
# 🦐 KRILL HERD LOOP
# =========================================
for itr in range(iterations):

    for i in range(num_krill):

        # Neighbor-induced movement
        neighbor_effect = np.mean(positions, axis=0) - positions[i]

        # Foraging motion (towards best)
        foraging = best_solution - positions[i]

        # Random diffusion
        diffusion = np.random.uniform(-1, 1, num_features)

        # Update position
        new_position = positions[i] + 0.5 * neighbor_effect + 0.5 * foraging + 0.1 * diffusion
        new_position = np.clip(new_position, 0, 1)

        new_fitness = fitness(new_position)

        if new_fitness > fitness_vals[i]:
            positions[i] = new_position
            fitness_vals[i] = new_fitness

        if new_fitness > best_score:
            best_solution = new_position
            best_score = new_fitness

    print(f"Iteration {itr+1} Best R2: {best_score:.4f}")

print("\n🏆 KA Best Score:", best_score)

# =========================================
# 🛟 FALLBACK
# =========================================
if best_solution is None:
    best_solution = np.ones(num_features)

# =========================================
# 🌳 FINAL MODEL
# =========================================
X_train_ka = X_train_scaled * best_solution
X_test_ka = X_test_scaled * best_solution

model = RandomForestRegressor(n_estimators=200, random_state=42)
model.fit(X_train_ka, y_train)

preds = model.predict(X_test_ka)

# =========================================
# 📈 METRICS
# =========================================
def safe_metrics(y_true, y_pred):
    try:
        return {
            "R2": float(r2_score(y_true, y_pred)),
            "RMSE": float(np.sqrt(mean_squared_error(y_true, y_pred))),
            "MAE": float(mean_absolute_error(y_true, y_pred))
        }
    except:
        return {"R2": 0, "RMSE": 0, "MAE": 0}

metrics = safe_metrics(y_test, preds)

print("\n📊 KA Metrics:", metrics)

# =========================================
# 💾 SAVE FILES
# =========================================
joblib.dump(model, os.path.join(save_dir, "ka_model.pkl"))
joblib.dump(scaler, os.path.join(save_dir, "ka_scaler.pkl"))

pd.DataFrame([metrics]).to_csv(os.path.join(save_dir, "ka_results.csv"), index=False)

pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": preds
}).to_csv(os.path.join(save_dir, "ka_predictions.csv"), index=False)

# =========================================
# 📊 GRAPHS
# =========================================

# Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(), annot=True)
plt.title("KA Heatmap")
plt.savefig(os.path.join(save_dir, "ka_heatmap.png"))
plt.show()

# Accuracy
plt.figure()
plt.bar(["R2"], [metrics["R2"]])
plt.title("KA Accuracy")
plt.savefig(os.path.join(save_dir, "ka_accuracy_graph.png"))
plt.show()

# Comparison
plt.figure()
plt.bar(["RMSE", "MAE"], [metrics["RMSE"], metrics["MAE"]])
plt.title("KA Error Comparison")
plt.savefig(os.path.join(save_dir, "ka_comparison_graph.png"))
plt.show()

# Prediction
plt.figure()
plt.plot(y_test.values, label="Actual")
plt.plot(preds, label="Predicted")
plt.legend()
plt.title("KA Prediction")
plt.savefig(os.path.join(save_dir, "ka_prediction_graph.png"))
plt.show()

# Result graph
plt.figure()
plt.bar(metrics.keys(), metrics.values())
plt.title("KA Result Metrics")
plt.savefig(os.path.join(save_dir, "ka_result_graph.png"))
plt.show()

# =========================================
# 📄 JSON + YAML
# =========================================
with open(os.path.join(save_dir, "ka_results.json"), "w") as f:
    json.dump(metrics, f, indent=4)

with open(os.path.join(save_dir, "ka_config.yaml"), "w") as f:
    yaml.dump({
        "target": target,
        "features": list(X.columns),
        "KA_best_score": float(best_score)
    }, f)

print("\n✅ KA PIPELINE COMPLETED SUCCESSFULLY")
```
