```python
# =========================================
# 📦 IMPORTS
# =========================================
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# =========================================
# 📁 PATHS
# =========================================
csv_path = r"C:\Users\NXTWAVE\Downloads\Food Subsidy Optimization & Forecasting System\Dataset.csv"
save_dir = r"C:\Users\NXTWAVE\Downloads\Food Subsidy Optimization & Forecasting System"
os.makedirs(save_dir, exist_ok=True)

# =========================================
# 📊 LOAD & CLEAN
# =========================================
df = pd.read_csv(csv_path)

for col in df.columns:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(",", "")
        .str.replace("₹", "")
        .str.strip()
    )
    df[col] = pd.to_numeric(df[col], errors='coerce')

df = df.dropna(axis=1, how='all')
df = df.fillna(df.mean())

if df.shape[1] < 2:
    raise Exception("❌ Need at least 2 numeric columns")

# =========================================
# 🎯 TARGET
# =========================================
target = df.columns[-1]
X = df.drop(columns=[target])
y = df[target]

# =========================================
# 🔀 SPLIT (SAFE FOR SMALL DATA)
# =========================================
test_size = 0.25 if len(df) > 4 else 0.5

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=test_size, random_state=42
)

# =========================================
# ⚙️ SCALING
# =========================================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# =========================================
# 🧬 AIS FITNESS (SAFE VERSION)
# =========================================
def fitness(weights):
    try:
        weights = np.array(weights)
        X_w = X_train_scaled * weights

        model = RandomForestRegressor(n_estimators=50, random_state=42)
        model.fit(X_w, y_train)

        preds = model.predict(X_test_scaled * weights)

        score = r2_score(y_test, preds)

        # Handle invalid values
        if np.isnan(score) or np.isinf(score):
            return -1e6

        return score

    except:
        return -1e6

# =========================================
# 🧬 AIS INITIALIZATION
# =========================================
num_features = X_train.shape[1]
population_size = 10
generations = 20

population = [np.random.rand(num_features) for _ in range(population_size)]

# 🔥 IMPORTANT FIX: initialize default solution
best_solution = np.ones(num_features)
best_score = -1e6

# =========================================
# 🧬 AIS LOOP
# =========================================
for gen in range(generations):
    scores = [fitness(ind) for ind in population]

    best_idx = np.argmax(scores)

    if scores[best_idx] > best_score:
        best_score = scores[best_idx]
        best_solution = population[best_idx]

    # Mutation
    new_population = []
    for ind in population:
        mutation = np.random.normal(0, 0.1, size=num_features)
        clone = np.clip(ind + mutation, 0, 1)
        new_population.append(clone)

    population = new_population

    print(f"Gen {gen+1} Best R2: {best_score:.4f}")

print("\n🏆 Best AIS Score:", best_score)

# =========================================
# 🛟 FALLBACK (CRITICAL FIX)
# =========================================
if best_solution is None:
    print("⚠️ AIS failed → using default weights")
    best_solution = np.ones(num_features)

# =========================================
# 🌳 FINAL MODEL
# =========================================
X_train_ais = X_train_scaled * best_solution
X_test_ais = X_test_scaled * best_solution

model = RandomForestRegressor(n_estimators=200, random_state=42)
model.fit(X_train_ais, y_train)

preds = model.predict(X_test_ais)

# =========================================
# 📈 METRICS
# =========================================
def safe_metrics(y_true, y_pred):
    try:
        return {
            "R2": float(r2_score(y_true, y_pred)),
            "RMSE": float(np.sqrt(mean_squared_error(y_true, y_pred))),
            "MAE": float(mean_absolute_error(y_true, y_pred))
        }
    except:
        return {"R2": 0, "RMSE": 0, "MAE": 0}

metrics = safe_metrics(y_test, preds)

print("\n📊 AIS Metrics:", metrics)

# =========================================
# 💾 SAVE FILES
# =========================================
joblib.dump(model, os.path.join(save_dir, "ais_model.pkl"))
joblib.dump(scaler, os.path.join(save_dir, "ais_scaler.pkl"))

pd.DataFrame([metrics]).to_csv(os.path.join(save_dir, "ais_results.csv"), index=False)

pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": preds
}).to_csv(os.path.join(save_dir, "ais_predictions.csv"), index=False)

# =========================================
# 📊 GRAPHS
# =========================================

# Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(), annot=True)
plt.title("AIS Heatmap")
plt.savefig(os.path.join(save_dir, "ais_heatmap.png"))
plt.show()

# Accuracy
plt.figure()
plt.bar(["R2"], [metrics["R2"]])
plt.title("AIS Accuracy")
plt.savefig(os.path.join(save_dir, "ais_accuracy_graph.png"))
plt.show()

# Comparison
plt.figure()
plt.bar(["RMSE", "MAE"], [metrics["RMSE"], metrics["MAE"]])
plt.title("AIS Error Comparison")
plt.savefig(os.path.join(save_dir, "ais_comparison_graph.png"))
plt.show()

# Prediction
plt.figure()
plt.plot(y_test.values, label="Actual")
plt.plot(preds, label="Predicted")
plt.legend()
plt.title("AIS Prediction")
plt.savefig(os.path.join(save_dir, "ais_prediction_graph.png"))
plt.show()

# Result graph
plt.figure()
plt.bar(metrics.keys(), metrics.values())
plt.title("AIS Result Metrics")
plt.savefig(os.path.join(save_dir, "ais_result_graph.png"))
plt.show()

# =========================================
# 📄 JSON + YAML
# =========================================
with open(os.path.join(save_dir, "ais_results.json"), "w") as f:
    json.dump(metrics, f, indent=4)

with open(os.path.join(save_dir, "ais_config.yaml"), "w") as f:
    yaml.dump({
        "target": target,
        "features": list(X.columns),
        "AIS_best_score": float(best_score)
    }, f)

print("\n✅ AIS FIXED PIPELINE COMPLETED")
```
