```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# =========================================
# 📁 PATHS
# =========================================
csv_path = r"C:\Users\NXTWAVE\Downloads\Food Subsidy Optimization & Forecasting System\Dataset.csv"
save_dir = r"C:\Users\NXTWAVE\Downloads\Food Subsidy Optimization & Forecasting System"
os.makedirs(save_dir, exist_ok=True)

# =========================================
# 📊 LOAD & CLEAN DATA
# =========================================
df = pd.read_csv(csv_path)

for col in df.columns:
    df[col] = (
        df[col].astype(str)
        .str.replace(",", "")
        .str.replace("₹", "")
        .str.strip()
    )
    df[col] = pd.to_numeric(df[col], errors='coerce')

df = df.dropna(axis=1, how='all')
df = df.fillna(df.mean())

if df.shape[1] < 2:
    raise Exception("❌ Need at least 2 numeric columns")

# =========================================
# 🎯 TARGET
# =========================================
target = df.columns[-1]
X = df.drop(columns=[target])
y = df[target]

# =========================================
# 🔀 SPLIT
# =========================================
test_size = 0.5 if len(df) <= 5 else 0.2

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=test_size, random_state=42
)

# =========================================
# ⚙️ SCALING
# =========================================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# =========================================
# 🧬 FITNESS FUNCTION
# =========================================
def fitness(position):
    try:
        position = np.array(position)
        X_w = X_train_scaled * position

        model = RandomForestRegressor(n_estimators=50, random_state=42)
        model.fit(X_w, y_train)

        preds = model.predict(X_test_scaled * position)
        score = r2_score(y_test, preds)

        if np.isnan(score) or np.isinf(score):
            return -1e6

        return score
    except:
        return -1e6

# =========================================
# 🐜 ALOA PARAMETERS
# =========================================
num_agents = 10
num_features = X_train.shape[1]
iterations = 20

# Initialize antlions and ants
antlions = np.random.rand(num_agents, num_features)
ants = np.random.rand(num_agents, num_features)

antlion_fitness = np.array([fitness(a) for a in antlions])

elite_idx = np.argmax(antlion_fitness)
elite = antlions[elite_idx]
elite_score = antlion_fitness[elite_idx]

# =========================================
# 🐜 ALOA LOOP
# =========================================
for itr in range(iterations):
    for i in range(num_agents):

        # Random walk around selected antlion
        rand_idx = np.random.randint(num_agents)
        selected_antlion = antlions[rand_idx]

        walk = np.random.uniform(-1, 1, num_features)
        new_ant = selected_antlion + walk

        # Move towards elite
        new_ant = (new_ant + elite) / 2

        new_ant = np.clip(new_ant, 0, 1)

        new_fitness = fitness(new_ant)

        # Replace antlion if better
        if new_fitness > antlion_fitness[rand_idx]:
            antlions[rand_idx] = new_ant
            antlion_fitness[rand_idx] = new_fitness

        # Update elite
        if new_fitness > elite_score:
            elite = new_ant
            elite_score = new_fitness

    print(f"Iteration {itr+1} Best R2: {elite_score:.4f}")

print("\n🏆 ALOA Best Score:", elite_score)

# =========================================
# 🛟 FALLBACK
# =========================================
if elite is None:
    elite = np.ones(num_features)

# =========================================
# 🌳 FINAL MODEL
# =========================================
X_train_aloa = X_train_scaled * elite
X_test_aloa = X_test_scaled * elite

model = RandomForestRegressor(n_estimators=200, random_state=42)
model.fit(X_train_aloa, y_train)

preds = model.predict(X_test_aloa)

# =========================================
# 📈 METRICS
# =========================================
def safe_metrics(y_true, y_pred):
    try:
        return {
            "R2": float(r2_score(y_true, y_pred)),
            "RMSE": float(np.sqrt(mean_squared_error(y_true, y_pred))),
            "MAE": float(mean_absolute_error(y_true, y_pred))
        }
    except:
        return {"R2": 0, "RMSE": 0, "MAE": 0}

metrics = safe_metrics(y_test, preds)

print("\n📊 ALOA Metrics:", metrics)

# =========================================
# 💾 SAVE FILES
# =========================================
joblib.dump(model, os.path.join(save_dir, "aloa_model.pkl"))
joblib.dump(scaler, os.path.join(save_dir, "aloa_scaler.pkl"))

pd.DataFrame([metrics]).to_csv(os.path.join(save_dir, "aloa_results.csv"), index=False)

pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": preds
}).to_csv(os.path.join(save_dir, "aloa_predictions.csv"), index=False)

# =========================================
# 📊 GRAPHS
# =========================================

# Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(), annot=True)
plt.title("ALOA Heatmap")
plt.savefig(os.path.join(save_dir, "aloa_heatmap.png"))
plt.show()

# Accuracy
plt.figure()
plt.bar(["R2"], [metrics["R2"]])
plt.title("ALOA Accuracy")
plt.savefig(os.path.join(save_dir, "aloa_accuracy_graph.png"))
plt.show()

# Comparison
plt.figure()
plt.bar(["RMSE", "MAE"], [metrics["RMSE"], metrics["MAE"]])
plt.title("ALOA Error Comparison")
plt.savefig(os.path.join(save_dir, "aloa_comparison_graph.png"))
plt.show()

# Prediction
plt.figure()
plt.plot(y_test.values, label="Actual")
plt.plot(preds, label="Predicted")
plt.legend()
plt.title("ALOA Prediction")
plt.savefig(os.path.join(save_dir, "aloa_prediction_graph.png"))
plt.show()

# Result graph
plt.figure()
plt.bar(metrics.keys(), metrics.values())
plt.title("ALOA Result Metrics")
plt.savefig(os.path.join(save_dir, "aloa_result_graph.png"))
plt.show()

# =========================================
# 📄 JSON + YAML
# =========================================
with open(os.path.join(save_dir, "aloa_results.json"), "w") as f:
    json.dump(metrics, f, indent=4)

with open(os.path.join(save_dir, "aloa_config.yaml"), "w") as f:
    yaml.dump({
        "target": target,
        "features": list(X.columns),
        "ALOA_best_score": float(elite_score)
    }, f)

print("\n✅ ALOA PIPELINE COMPLETED SUCCESSFULLY")
```
